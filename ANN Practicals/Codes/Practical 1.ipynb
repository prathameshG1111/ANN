```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10)
plt.xlabel("Input")
plt.ylabel("Output")
plt.plot(x, x, label='Linear')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(10,-10,100)
y = 1/(1+np.exp(-x))
plt.xlabel("Input") 
plt.ylabel("Sigmoid(X)") 
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.xlabel("Input") 
plt.ylabel("Tanh(X)") 
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, np.maximum(0, x))
plt.xlabel("Input")
plt.ylabel("ReLU(x)")
plt.grid(True)
plt.show()
```

```python
def softmax(x):
    exp_values = np.exp(x - np.max(x))
    return exp_values / np.sum(exp_values)

x = np.linspace(-5, 5, 100)
y = softmax(x)

plt.xlabel("Input")
plt.ylabel("Output")
plt.plot(x, y)
plt.title("Softmax Activation Function")
plt.grid(True)
plt.show()
```
